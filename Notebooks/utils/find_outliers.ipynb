# Removing Outliers

Not all data is clean.
Most of the times, what we receive is noisy data.

Even among noisy datapoints, some of them can be easily seen to be too far off the general data manifold to make sense.

These are the _anomalies_ or the _outliers_.

For example, say there are a bunch of MNIST digits we want to train on. Few of the images (a handful) are just bad memes (I wonder how one would have a low res, 28x28 meme, though!).

If we do anything with the data without removing these outliers, they may be detrimental to our calculations (say, calculating the mean).

So, as routine cleanup, we would like to have an algorithm find these outliers for us.

In the MNIST example above, the data is high (784) dimensional.
So, one needs to have an abstract idea of what an outlier looks like and where the data manifold is.

This is not our task at present.

As a first step (a very important one), we would like to clean up a signal based on simple metrics, like the distance between the datapoints.

According to http://www.itl.nist.gov/div898/handbook/eda/section3/eda35h.htm,
a better way to remove outliers is to calculate the median (instead of the mean) and the Euclidean distance (instead of the standard deviation).

Given a set of n-dimensional data points $\textbf{x}$ shaped $(-1, n)$, we first find the median $\bar{\textbf{x}}$ and the distances $d$ between $\textbf{x}$ and $\bar{\textbf{x}}$
That is,
$$d = \lVert \textbf{x} - \bar{\textbf{x}} \rVert$$

Also, the median of d is found as $\bar{d}$.

Then, _the modified Z-score_ is defined as:
$$s = \gamma \frac{d}{\bar{d}}$$

where $\gamma = 0.6745$ is a constant.

The outliers are those points for which $s > \beta$, where $\beta$ is a _threshold_, usually set to 3.5.

Let's implement the basic functionality

## Define Function

````python
def _find_outliers(x, threshold=3.5):
    GAMMA = 0.6745
    
    x_bar = np.median(x, 0)
    d = np.linalg.norm(x - x_bar, axis=-1)
    d_bar = np.median(d)
    
    s = GAMMA * d / d_bar

    outlier_mask = s > threshold
    
    return outlier_mask
````

## Test

Let's first define a simple convinience function

````python
from vai.utils import __spline_interpolate as __interpolate
````

````python
def __make_outliers(y, outlier_mask, y_lim):
    n = len(np.where(outlier_mask)[0])
    
    upper_range = (y[outlier_mask] + y_lim[1]) / 2
    upper_array = upper_range + np.random.rand(n) * (y_lim[1] - upper_range)
    lower_range = (y[outlier_mask] + y_lim[0]) / 2
    lower_array = y_lim[0] + np.random.rand(n) * (lower_range - y_lim[0])
    probs = np.stack([y_lim[1] - y[outlier_mask], y[outlier_mask] - y_lim[0]], -1)
    probs /= probs.sum(-1, keepdims=True)
    choice_idx = np.concatenate([np.random.binomial(1, prob[1], 1) for prob in probs])
    y[outlier_mask] = np.stack([upper_array, lower_array], -1)[np.arange(n), choice_idx]
````

# Modified Function

In order to remedy this, we need to focus!

We need to split the data into equal chunks/windows and then do the same thing.

This, allows the algorithm to remove all _local_ outliers as well.

Let's do this.

````python
def find_outliers(x, threshold=3.5, window_fraction=0.05):
    window_length = int(len(x) * window_fraction)
    divide_ids = np.arange(window_length, len(x), window_length)

    split_data = np.stack([x[i:i + window_length] for i in range(len(x) - window_length)])
    
    outlier_idx = np.concatenate([np.arange(i, i + window_length)[_find_outliers(d, threshold)] for i, d in enumerate(split_data)])
    
    return np.array([i in np.unique(outlier_idx) for i in range(len(x))])
````

````python
def test_filter(fn=lambda x: 3 - np.sqrt(x), x_lim=(0, 9), y_lim=None,
                n=100, frac_outlier=0.1, std=10):
    import ipywidgets as widgets
    %matplotlib notebook
    
    x = np.linspace(x_lim[0], x_lim[1], n)
    y = fn(x)
    y += np.random.randn(n) * y / std
    
    fig = plt.figure(figsize=(12, 8))
    ax = fig.add_subplot(111)
    
    if y_lim is None: y_lim = (y.min(), y.max())
    
    # Make some points outliers
    outlier_idx = np.random.choice(np.arange(n), int(n * frac_outlier), replace=False)
    outlier_mask = np.array([i in outlier_idx for i in range(n)])
    __make_outliers(y, outlier_mask, y_lim)
    
    ax.plot(x[outlier_mask], y[outlier_mask], 'or')
    ax.plot(x[~outlier_mask], y[~outlier_mask], 'og')
    
    def update(threshold=3.5, frac=0.05):
        if threshold != update.threshold or frac != update.frac:
            outlier_mask = find_outliers(np.expand_dims(y, -1), threshold, frac)
            new_y = y.copy()
            new_y[outlier_mask] = __interpolate(x[~outlier_mask], y[~outlier_mask], x[outlier_mask])
            line.set_ydata(new_y)
            update.threshold = threshold
            update.frac = frac
    
    update.threshold = 3.5
    update.frac = 0.05
    outlier_mask = find_outliers(np.expand_dims(y, -1), update.threshold, update.frac)
    new_y = y.copy()
    new_y[outlier_mask] = __interpolate(x[~outlier_mask], y[~outlier_mask], x[outlier_mask])
    line, = ax.plot(x, new_y, '--b')
    
    widgets.interact(update, threshold=(0, 5, 0.5), frac=(0, 1, 0.01))
````
